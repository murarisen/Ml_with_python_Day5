### Remove Outliers

```python
def remove_outlier(col):
    sorted(col)
    q1,q3=col.quantile([0.25,0.75])
    iqr=q3-q1
    lower_range=q1-(1.5*iqr)
    upper_range=q3+(1.5*iqr)
    return lower_range,upper_range
```

```python
lowleadtime,uppleadtime=remove_outlier(mydata['lead_time'])
mydata['lead_time']=np.where(mydata['lead_time']>uppleadtime,uppleadtime,mydata['lead_time'])
mydata['lead_time']=np.where(mydata['lead_time']<lowleadtime,lowleadtime,mydata['lead_time'])
```
